# Import thư viện

```python
import math
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score as r2, mean_squared_error as mse
```

# Hàm cơ sở

```python
# cost or loss function
def cost(Y, Yhat):
  return np.sum(np.linalg.norm(Yhat - Y))
```

# Bài 1

```python
df = pd.read_csv('/home/vitquay1708/Study_Space/ml/week_9/data/SAT_GPA.csv')
```

```python
X = df['SAT'].to_numpy()
y = df['GPA'].to_numpy()

X = X.reshape(-1, 1)
y = y.reshape(-1, 1)

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=54, random_state=42)
y_train = y_train.reshape(-1, 1)
y_test = y_test.reshape(-1, 1)
```

## Khởi tạo giá trị

```python
d0 = 1
d1 = h = 100 # size of hidden layer

d2 = 1 #output size
# initialize parameters randomly
W1 = 0.01*np.random.randn(d0, d1)
b1 = np.zeros((d1, 1))
W2 = 0.01*np.random.randn(d1, d2)
b2 = np.zeros((d2, 1))
print('W1: ', W1.shape)
print('X_train:', X_train.shape)
print('b1:',b1.shape)
print('w2:', W2.shape)
print('b2',b2.shape)

N = X_train.shape[0]
eta = 1 # learning rate
```

## Feedforward và backpropagation

```python
for i in range(10000):
## Feedforward
  Z1 = np.dot(X_train, W1) + b1.T
  A1 = np.maximum(Z1, 0)
  Z2 = np.dot(A1, W2) + b2
  Yhat = Z2
  # print loss after each 1000 iterations
  if i %1000 == 0:
    # compute the loss: average cross-entropy loss
    loss = cost(y_train, Yhat)
    print("iter %d, loss: %f" %(i, loss))

  # backpropagation
  E2 = (Yhat - y_train) / N
  dW2 = np.dot(A1.T, E2)
  db2 = np.sum(E2, axis = 0, keepdims = True)
  E1 = np.dot(E2, W2.T)
  E1[Z1 <= 0] = 0 # gradient of ReLU
  dW1 = np.dot(X_train.T, E1)
  db1 = np.sum(E1, axis = 0, keepdims = True).T
  # Gradient Descent update
  W1 += -eta*dW1
  b1 += -eta*db1
  W2 += -eta*dW2
  b2 += -eta*db2
```

```python
Z1 = np.dot(W1.T, X_test.T) + b1
A1 = np.maximum(Z1, 0)
Z2 = np.dot(W2.T, A1) + b2
print(Z2.shape)
print(y_test.shape)
print(y_test.T[0])
print(Z2[0])
print(mse(y_test.T[0], Z2[0]))
print(r2(y_test.T[0], Z2[0]))
```
