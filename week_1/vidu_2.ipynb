```python
from sklearn.naive_bayes import MultinomialNB, BernoulliNB
import numpy as np
```

```python
d1 = [2, 1, 1, 0, 0, 0, 0, 0, 0]
d2 = [1, 1, 0, 1, 1, 0, 0, 0, 0]
d3 = [0, 1, 0, 0, 1, 1, 0, 0, 0]
d4 = [0, 1, 0, 0, 0, 0, 1, 1, 1]

train_data = np.array([d1, d2, d3, d4])
label = np.array(['B', 'B', 'B', 'N'])

d5 = np.array([[2, 0, 0, 1, 0, 0, 0, 1, 0]])
d6 = np.array([[0, 1, 0, 0, 0, 0, 0, 1, 1]])
```

```python
clf1 = MultinomialNB()

clf1.fit(train_data, label)
print('Predicting class of d5:', str(clf1.predict(d5)[0]))
print('Predicting class of d6:', str(clf1.predict(d6)[0]))
print('Probability of d5 in each class:', clf1.predict_proba(d5))
print('Probability of d6 in each class:', clf1.predict_proba(d6))
```

```python
clf2 = BernoulliNB()
clf2.fit(train_data, label)
# test
print('Predicting class of d5:', str(clf2.predict(d5)[0]))
print('Predicting class of d6:', str(clf2.predict(d6)[0]))
print('Probability of d5 in each class:', clf2.predict_proba(d5))
print('Probability of d6 in each class:', clf2.predict_proba(d6))
```
